```python
import wget
from PIL import Image
import os
import urllib
import glob
```

```python
def check_img_or_del(full_name):
    try:
        im=Image.open(full_name)
        print("It's an image")
    except IOError:
        print("It's NOT an image, removing whatever it is")
        os.remove(full_name)
```

```python
def download_multiple_times(url, full_name, counter):
    print("Try %s" %(counter))
    try:
        #wget.download(url, full_name)
        request = urllib.request.urlopen(url, timeout=10)
        with open(full_name, 'wb') as f:
            try:
                f.write(request.read())
            except:
                print("error")
        check_img_or_del(full_name)
        print("Downloaded successfully")
        return 0
    except:
        counter_new = counter - 1
        if counter == 1:
            print("Could not find the image %s" %(url))
        else:
            return download_multiple_times(url, full_name, counter_new)
```

```python
text_file = "ImageNet-Rocks.txt"
root_name = "./data/rck_"
tries = 2
```

```python
with open(text_file) as f:
    lines = f.readlines()
lines = [line.rstrip('\n') for line in lines]
```

```python
print ("Started downloading for file %s" %(text_file))

for i in range(len(lines)):  
    url = lines[i]
    print('Index %s, Downloading image %s' %(i, url))
    
    full_name = root_name + str(i).zfill(5) + ".jpg"
    download_multiple_times(url, full_name, tries)
    print("\n")

print("Finished Downloading %s images" %(len(lines)))
```
